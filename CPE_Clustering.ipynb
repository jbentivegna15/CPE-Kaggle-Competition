```python
# Ariana Freitag
# Clustering Assignment
# Dataset created for CPE Kaggle Competition

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans

df = pd.read_csv("./CPE DATASET.CSV")
df_nostring = df.drop(['Unnamed: 0', 'City'], axis=1)
df_changed = df_nostring.dropna(axis = 1)
```

```python
# cluster by census data
km = KMeans(n_clusters=3, random_state=2).fit_predict(df_changed)
cmap = cm.get_cmap('Spectral')

plt.scatter(df_changed.values[:,0], df_changed.values[:,1], c=km, cmap=cmap)
```

```python
city = df['City'].dropna()
df1 = df.dropna(axis=1)
df1['City'] = city
dfcity1 = pd.concat([df1,pd.DataFrame(km, dtype=np.int)], axis=1, ignore_index=True)
# can see on the table where the cities are grouped
```

```python
# cluster by race and gender
df2 = df_nostring[['Race_Native American','Race_Black', 'Race_Latinx', 'Race_Pacific Islander', 'Race_Not Listed', 'Race_White', 'Gender_Male', 'Gender_Female']].copy()
df2 = df2.dropna()
km2 = KMeans(n_clusters=3, random_state=2).fit_predict(df2)
df2['City'] = df['City'] 
plt.scatter(df2.values[:,0], df2.values[:,1], c=km2, cmap=cmap)
```

```python
# Use of Force clustering
df3 = df_nostring[['UOF_Use of Physical Force', 'UOF_Use of Non-Lethal Weapon', 'UOF_Use of Lethal Weapon']]
df3 = df3.dropna()
km3 = KMeans(n_clusters=3, random_state=2).fit_predict(df3)

plt.scatter(df3.values[:,0], df3.values[:,1], c=km3, cmap=cmap)
```

```python
# crime type clustering
df4 = df_nostring[['Crime_Violent', 'Crime_Non-Violent', 'Crime_Non-Criminal']]
df4 = df4.dropna()
km4 = KMeans(n_clusters=3, random_state=2).fit_predict(df4)

plt.scatter(df4.values[:,0], df4.values[:,1], c=km4, cmap=cmap)
```
